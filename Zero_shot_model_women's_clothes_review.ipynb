```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True) 
```

```python
df = df.dropna()
```

```python
df1 = df.sample(500)
```

Initiliazing the zero-shot facebook bart large mnli model

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
df1.head(5)
```

Provide the text/phrase to classifer and along with the rating labels to be classified.

Classifier provides the scores for labels that are matching to the text/phrase given as the input. 

Here in the below example we have considered Positive and Negative as the lables and text/phrase scores explains which label is close. 

In the below example, for text 'Yuck', it has classified as Negative with 99.3%

```python
sequence_to_classify = df1['Title'].values[0]
rating_labels = ['Positive', 'Negative']
classifier(sequence_to_classify, rating_labels)
```
